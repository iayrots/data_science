```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

##### Bar plot

```python
y = np.random.randint(10, size=5)
x = range(5)
plt.bar(x, y)
plt.show()
```

```python
y = np.random.randint(5, 10, size=5)
x = range(5) 
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.bar(x, y, yerr=error, align='center', alpha=0.5) # alpha : 투명도
plt.show()
```

```python
y = np.random.randint(5, 10, size=5)
x = range(5) 
error = np.random.randint(1, 5, size=5)
print(y, error)
plt.barh(x, y, xerr=error, align='center', alpha=0.6)  # barh : 눕혀줌
plt.show()
```

```python
y1 = np.random.randint(5, 10, size=5)  # 더 큰 데이터를 먼저 그려야함
y2 = np.random.randint(1, 5, size=5)
print(y1, y2)
plt.subplots()
x = range(5)
plt.bar(x, y1)
plt.bar(x, y2)
plt.tight_layout()  # layout을 꽉 채워줌. 생략가능
plt.show()
```

```python
y1 = np.random.randint(1, 5, size=5)
y2 = np.random.randint(1, 5, size=5)
bar_width, opacity = 0.4, 0.6
print(y1, y2, bar_width, opacity)
x = np.arange(5)

plt.bar(x, y1, bar_width, alpha=opacity, color='b', label='randint_1', align="edge")
plt.bar(x + bar_width, y2, bar_width, alpha=opacity, color='r', label='randint_2', align="edge")

plt.xlabel('X Label') # X 라벨
plt.ylabel('Value')   # Y 라벨
plt.xticks(x + bar_width, ('A', 'B', 'C', 'D', 'E')) # xtick 추가     bar_width로 x축값 가운데로 옮겨줌
plt.title('Random Integer') # 제목추가
plt.legend() # 범례추가
plt.show()
```

##### pie plot

```python
data = np.random.randint(1,6, size=5)
labels = ['A','B','C','D','E']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral', 'red']
print(data)
plt.pie(data, labels=labels, colors=colors, autopct='%1.1f%%') # f:float, g:그 자리에서 반올림, d:정수
plt.show()
```

##### sactter plot

```python
# uniform (균등분포)
x1 = np.random.rand(500)
y1 = np.random.rand(500)
plt.subplot(311)
plt.scatter(x1, y1)

# gaussian (정규분포)
x2 = np.random.randn(500)
y2 = np.random.randn(500)
plt.subplot(312)
plt.scatter(x2, y2)

# randint (uniform)
x3 = np.random.randint(1,100,size=500)
y3 = np.random.randint(1,100,size=500)
plt.subplot(313)
plt.scatter(x3, y3)

plt.show()
```

```python
data1 = np.random.randint(1,10, size=20)
data2 = np.random.randint(1,10, size=20)
colors = np.random.rand(20)
area = np.pi * np.arange(10,50,2) ** 2    # 원의 면적
plt.scatter(data1, data2, s=area, c=colors, alpha=0.5)
plt.show()


# N = 50
# x = np.random.rand(N)
# y = np.random.rand(N)

# area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

# plt.scatter(x, y, s=area, c=colors, alpha=0.5)
# plt.show()
```

##### histogram

```python
# 50개의 구간을로 잘라 해당되는 데이터의 수를 카운트한 후에 y축의 값을 normalized 한다
x1 = np.random.rand(1000)
plt.subplot(211)
plt.hist(x1, bins=50, normed=True)  # bins : 몇등분으로 나눌지

x2 = np.random.randn(1000)
plt.subplot(212)
plt.hist(x2, bins=50, normed=True)

plt.show()
```

```python
x = np.random.randint(0,100, size=100)
print(x, np.bincount(x))
plt.subplot(311)
plt.hist(x)
plt.subplot(312)
plt.hist(x, bins=100) 
plt.subplot(313)
plt.hist(x, normed=True)

plt.show()
```
